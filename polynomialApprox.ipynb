```python
from autograd import numpy as np
from matplotlib import pyplot as plt
```

```python
p = np.array([1, 2, -1, 0.5])
xs = np.arange(-5, 5, 0.001)
plt.plot(xs, np.polyval(p, xs))
plt.show()
```

```python
psin = np.array([1/362880, 0, -1/5040, 0, 1/120, 0, -1/6, 0, 1, 0])
xs = np.arange(-5, 5, 0.001)
plt.plot(xs, np.polyval(psin, xs))
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import special as sp
import itertools as it
```

```python
c = it.cycle([0.0, 1.0, 0.0, -1.0])
N = 50

p = 1.0 / sp.factorial(np.arange(N)) * np.array(list(it.islice(c, N)))
p = np.flip(p)

xs = np.arange(-5, 5, 0.001)
plt.plot(xs, np.polyval(p, xs))
plt.show()
```
